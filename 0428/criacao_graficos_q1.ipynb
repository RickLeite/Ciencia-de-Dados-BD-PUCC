### Criação de Gráficos em Python

##### 1) Você foi contratado por uma empresa de marketing para analisar o desempenho de vendas de dois produtos ao longo dos último ano. A empresa deseja visualizar as tendências de vendas de forma clara e impactante. Qual tipo de visualização você recomendaria? Justifique sua resposta e apresente um gráfico que represente a situação descrita.

#### Geração dos Dados

```python
import pandas as pd
import numpy as np

meses = pd.date_range(start='2023-01-01', end='2024-01-01', freq='BME') 
vendas_produto_A = np.random.randint(100, 500, size=len(meses))
vendas_produto_B = np.random.randint(100, 500, size=len(meses))

dados = pd.DataFrame({
    'Meses': meses,
    'Vendas_Produto_A': vendas_produto_A,
    'Vendas_Produto_B': vendas_produto_B
})

dados.head()
```

#### Gráfico Impactante

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")


plt.figure(figsize=(18, 6))


# Linha para o Produto A
sns.lineplot(data=dados, x='Meses', y='Vendas_Produto_A', label='Produto A', color='skyblue')

# Pontos de dados para o Produto A
sns.scatterplot(data=dados, x='Meses', y='Vendas_Produto_A', color='blue', s=100, marker='o')

# Linha para o Produto B
sns.lineplot(data=dados, x='Meses', y='Vendas_Produto_B', label='Produto B', color='orange')

# Pontos de dados para o Produto B
sns.scatterplot(data=dados, x='Meses', y='Vendas_Produto_B', color='red', s=100, marker='o')

# sombra às áreas abaixo das linhas
plt.fill_between(dados['Meses'], dados['Vendas_Produto_A'], color="skyblue", alpha=0.2)
plt.fill_between(dados['Meses'], dados['Vendas_Produto_B'], color="orange", alpha=0.2)

# título e rótulos dos eixos
plt.title('Desempenho de Vendas dos Produtos A e B ao Longo do Ano', fontsize=16, fontweight='bold')
plt.xlabel('Meses', fontsize=14)
plt.ylabel('Vendas', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# legenda
plt.legend(fontsize=12)

plt.show()
```

### Recomendado: Gráfico de Linhas com duas séries de dados

Cada linha representa as vendas de um produto específico ao longo do tempo (meses do ano). <br> <br>
As cores diferentes <font color="skyblue"> ------- </font> <font color="orange"> ------- </font> são usados para distinguir os dois produtos. <br> <br>
Facilitando a comparação entre as tendências de vendas de cada produto e padrões semelhantes ou divergentes ao longo do tempo
