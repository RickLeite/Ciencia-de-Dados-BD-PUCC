### Criação de Gráficos em Python

##### 7) (Categorias). Você está realizando uma pesquisa para identificar as preferências musicais dos estudantes em sua escola. Através de questionários preenchidos por 100 alunos, você coletou os seguintes dados sobre seus gêneros musicais favoritos:
- Rock: 35 alunos
- Pop: 25 alunos
- Hip Hop: 15 alunos
- Eletrônica: 10 alunos
- Jazz: 5 alunos
- Clássica: 10 alunos

#### Geração dos Dados

```python
import pandas as pd

generos_musicais = ["Rock", "Pop", "Hip Hop", "Eletrônica", "Jazz", "Clássica"]
frequencias = [35, 25, 15, 10, 5, 10]

dados = pd.DataFrame({'Gênero Musical': generos_musicais, 'Frequência': frequencias})
```

#### Gráfico Impactante

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.pie(dados['Frequência'], labels=dados['Gênero Musical'], autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})

plt.title('Preferências Musicais dos Alunos')

plt.show()
```
