#### 24) Aplique o cálculo da distância interquartílica para determinar se existem valores outliers na coluna "Gender Inequality Index (2021)" Lembre-se que o cálculo da distância interquartílica é dado por: 

IQR = q3 - q1

Limite_inferior = q1 - 1.5*IQR

Limite_superior = q3 + 1.5*IQR

```python
q1 = df['Gender Inequality Index (2021)'].quantile(0.25)
q3 = df['Gender Inequality Index (2021)'].quantile(0.75)
IQR = q3 - q1

limite_inferior = q1 - 1.5 * IQR
limite_superior = q3 + 1.5 * IQR

outliers = df[(df['Gender Inequality Index (2021)'] < limite_inferior) | (df['Gender Inequality Index (2021)'] > limite_superior)]
outliers
```

Correlação forte entre IDH Rank (2021) e Gender Inequality Index (2021), <br> Irei utiliza-la para predizer os valores faltantes do Gender Inequality Index (2021)

```python

# Preparando os dados
train_data = df[df['Gender Inequality Index (2021)'].notnull() & df['IDH Rank (2021)'].notnull()]
test_data = df[df['Gender Inequality Index (2021)'].isnull() & df['IDH Rank (2021)'].notnull()]

# Treinando o modelo
model = LinearRegression()
model.fit(train_data[['IDH Rank (2021)']], train_data['Gender Inequality Index (2021)'])

# Predizendo valores faltantes
predicted_values = model.predict(test_data[['IDH Rank (2021)']])
df.loc[df['Gender Inequality Index (2021)'].isnull() & df['IDH Rank (2021)'].notnull(), 'Gender Inequality Index (2021)'] = predicted_values
```

```python
# Usando modelo que realizei da ultima atividade

quartiles = df['Gender Inequality Index (2021)'].quantile([0.25, 0.5, 0.75])
q1 = quartiles[0.25]
q2 = quartiles[0.5]
q3 = quartiles[0.75]

plt.figure(figsize=(8,6))
plt.boxplot(df['Gender Inequality Index (2021)'], vert=False)
plt.xlabel('Gender Inequality Index (2021)')
plt.title('Boxplot Gender Inequality Index (2021)')

legenda_quartis  = [f'Q1: {q1}', 
                 f'Q2 (Mediana): {q2}', 
                 f'Q3: {q3}']
plt.legend(legenda_quartis , loc='upper right')

plt.yticks([])
plt.grid(True)
plt.show()
```

#### 28) Faça um gráfico de linha que apresente duas séries:

    -A primeira representa a  evolução do Brasil no período de 1990 a 2021
    -A segunda série representa a média do indice dos paises da LAC.
    
    Eixo X - Anos
    Eixo Y - Índice.

```python
dados_temporais = df.filter(regex=r'^Gender Inequality Index \(\d{4}\)$').copy()
dados_temporais['ISO3'] = df['ISO3']
dados_temporais
```

```python
dados_temporais.set_index('ISO3', inplace=True)
df_pivotado = dados_temporais.melt(var_name='Ano', value_name='Índice de Desigualdade de Gênero', ignore_index=False).reset_index()
df_pivotado['Ano'] = df_pivotado['Ano'].str.extract(r'\((\d{4})\)')
df_pivotado['Ano'] = pd.to_numeric(df_pivotado['Ano'])
df_pivotado = df_pivotado.sort_values(by='Ano')
df_pivotado.head()
```

```python
df_pivotado_join_undp = df_pivotado.join(df[['ISO3', 'Regiões em Desenvolvimento do UNDP']].set_index('ISO3'), on='ISO3')
```

```python
df_pivotado_join_undp.head()
```

```python
df_indice_br = df_pivotado_join_undp[df_pivotado_join_undp['ISO3'] == 'BRA']
df_indice_br.head()
```

```python
indice_br_series = df_indice_br.set_index('Ano')['Índice de Desigualdade de Gênero']
indice_br_series.head()
```

```python
regiao_lac = df_pivotado_join_undp[df_pivotado_join_undp['Regiões em Desenvolvimento do UNDP'] == 'LAC']
lac_media_ano = regiao_lac.groupby('Ano')['Índice de Desigualdade de Gênero'].mean()
lac_media_ano.head()
```

```python

plt.figure(figsize=(10, 6))
plt.plot(indice_br_series.index, indice_br_series.values, marker='o', linestyle='-', label='Brasil')
plt.plot(lac_media_ano.index, lac_media_ano.values, marker='s', linestyle='--', label='Média Países LAC')
plt.title('Índice de Desigualdade de Gênero ao longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Índice de Desigualdade de Gênero')
plt.legend()
plt.grid(True)
plt.show()
```

#### 29) Escolha um país a sua escolha e faça um gráfico de linha para apresentar a evolução do índice (Similar ao Brasil no exercício anterior). A diferença que desta vez os dados deverão estar rotulados. Dica: Utilize a função Annotate.

```python
df_indice_netherlands = df_pivotado[df_pivotado['ISO3'] == 'NLD']
df_indice_netherlands.head()
indice_nld_series = df_indice_netherlands.set_index('Ano')['Índice de Desigualdade de Gênero']
indice_nld_series.head()
```
